Well, some of these columns have mostly one value and are therefore candiddats to be dropped.
The 'nrOfPictures' column needs to be investigated some more. Some of the numeric data in the set has been stored as text data and will need to be cleaned.


```python
autos = autos.drop(["num_photos", "seller", "offer_type"], axis=1)
```

```python
autos['price']=autos['price'].str.replace('$','').str.replace(',','').astype(float)
```

```python
autos['odometer']=autos['odometer'].str.replace('km','').str.replace(',','').astype(float)
autos['odometer']
```

```python
autos.rename(columns={'odometer':'odometer_km'},inplace=True)
```

Here I removed all rows where the preice of the vehicle was less than one. I found that there were a lot of zero values and I had to clean them. Next, I will clean te 'odometer_km' column

```python
autos[(autos['registration_year']<1900) | (autos['registration_year']>2016)].describe()
```

```python
autos=autos[(autos['registration_year']>1900) & (autos['registration_year']<2016)]
```

```python
autos['registration_year'].value_counts(normalize=True)
```

```python
brand_counts = autos["brand"].value_counts(normalize=True)
brand_counts
```

```python
common_brands = brand_counts[brand_counts > .05].index
print(common_brands)
```

```python
brand_mean_prices = {}

for brand in common_brands:
    brand_only=autos[(autos['brand']==brand)]
    mean_prices=brand_only['price'].mean()
    brand_mean_prices[brand]=int(mean_prices)
    
brand_mean_prices
```

```python
brand_mean_mileage={}

for brand in common_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_mileage = brand_only["odometer_km"].mean()
    brand_mean_mileage[brand] = int(mean_mileage)
    
    
mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending=False)
mean_prices = pd.Series(brand_mean_prices).sort_values(ascending=False)
    
```

```python
brand_info = pd.DataFrame(mean_mileage,columns=['mean_mileage'])
brand_info
```

```python
brand_info['mean_prices']=mean_prices
brand_info
```


The range of car mileages does not vary as much as the prices do by brand, instead all falling within 10% for the top brands. There is a slight trend to the more expensive vehicles having higher mileage, with the less expensive vehicles having lower mileage.
